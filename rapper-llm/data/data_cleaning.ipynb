```python
# cleaning songs

import re

def clean_lyrics(lyrics):
    # Remove the unwanted text at the start before the first "[" and "Embed" at the end
    # Assumes that the unwanted starting text does not contain the "[" character
    cleaned_lyrics = re.sub(r"^[^\[]*\[", "[", lyrics)  # Remove text before the first "["
    cleaned_lyrics = re.sub(r"\d+Embed.*", "", cleaned_lyrics) # Remove text at the end

    # Remove content inside parentheses
    cleaned_lyrics = re.sub(r"\(.*?\)", "", cleaned_lyrics)

    # Additional cleanup: convert to lowercase
    cleaned_lyrics = cleaned_lyrics.lower()

    # Replace multiple spaces with a single space
    cleaned_lyrics = re.sub(r"\s+", " ", cleaned_lyrics).strip()

    return cleaned_lyrics

# Sample input to test
lyrics_sample = """"170 ContributorsHate Bein’ Sober Lyrics[Intro: Chief Keef] Young Chop on the beat I can't spell sober I'm a smoker, Fredo a drinker, Tadoe off molly water Sober, my bitches stay sober Sober Damn, I hate bein' sober I hate bein' sober [Chorus: Chief Keef] Damn, I hate bein' sober, I'm a smoker Fredo a drinker, Tadoe off molly water But we can't spell sober Ballout roll up, when we roll up, bitches be on us All the hoes, they love smokin' and love drinkin' Anti-sober for no reason 'Cause we can't spell sober Y'all know us, we smoke strong, bruh Watch me roll up 'Cause I can't spell sober [Verse 1: Chief Keef] On my tour bus we get dumb high, you's a floor boy Fredo got a hangover, he totin' a Cobra Last night he was shootin' shit up like O-Dog Reese rollin', Tadoe got hoes on mollies Chief Sosa, Ballout, we high ridin' 'Raris My bitches love drinkin', some love smokin' Let my alcoholic bitch hit the dutch, she start chokin' Call up D-Money, now we throw money All these bitches off the shits walk around like some zombies Call up D-Money, now we throw money All these bitches off the shits walk around like some zombies, yeah You might also like[Chorus: Chief Keef] Damn, I hate bein' sober, I'm a smoker Fredo a drinker, Tadoe off molly water But we can't spell sober Ballout roll up, when we roll up, bitches be on us All the hoes, they love smokin' and love drinkin' Anti-sober for no reason 'Cause we can't spell sober Y'all know us, we smoke strong, bruh Watch me roll up 'Cause I can't spell sober [Verse 2: 50 Cent] We got a hundred pounds of this shit (Yeah) My stash house with them bricks (Yeah) My pockets filled with them stacks (Uh) My bitch be gone off a flat (Whoo) She a hot tamale when she pop a molly It's time to party, we party hard Drinkin', smokin', drinkin', smokin', drinkin', smokin' We high for sure I came in back of that Rolls Nigga, I ain’t stuntin' them hoes (Nah) I trick a bitch to suck dick, trick What you spend on her, we spend on clothes Too young for me, she want Sosa Shooters in the Range Rover That's GBE, when them .223's get to flyin', bitch, it's over See my ring chain and my Rolex when I’m flexin' Bitch, I got to get mine, nigga get outta line, I check 'em See this gangsta shit's done to perfection Nigga, better believe me, I make it look easy [Chorus: Chief Keef] Damn, I hate bein' sober, I'm a smoker Fredo a drinker, Tadoe off molly water But we can't spell sober Ballout roll up, when we roll up, bitches be on us All the hoes, they love smokin' and love drinkin' Anti-sober for no reason 'Cause we can't spell sober Y'all know us, we smoke strong, bruh Watch me roll up 'Cause I can't spell sober [Verse 3: Wiz Khalifa] Uh, uh, uh My weed so strong, my cheese so long (My cheese so long) Roll so many joints, soon I might need a lung (Might need a lung) Spend so many grands that I might need some bands (Might need some bands) That's your bitch, why she actin' like she need a man? (Need a man) I’m faded (Faded), talkin' mill's 'cause I made it (Made it) Talkin' pounds 'cause I smoke it (Smoke it) Talkin' game 'cause I played it (I played it) I’m wasted (Wasted), Rose thats my favorite (Favorite) OG kush, you can taste it (Taste it) Buying Cris' by the cases (By the cases) I hate being sober Don't smell no one smokin', me and my niggas gon' roll up Believe they gon' fire on you You think you could roll up You smoke by the ounce Well, bitch, I smoke by the pound 'cause [Chorus: Chief Keef & Wiz Khalifa] Damn I hate bein' sober (Hahaha), I'm a smoker Fredo a drinker (Uh), Tadoe off molly water But we can't spell sober (Sosa, what up nigga?) Ballout roll up (What up Fredo?), when we roll up, bitches be on us All the hoes, they love smokin' and love drinkin' Anti-sober for no reason 'Cause we can't spell sober Y'all know us, we smoke strong, bruh Watch me roll up 'Cause I can't spell sober114Embed"
"""

# Clean the lyrics
cleaned_lyrics = clean_lyrics(lyrics_sample)
print(cleaned_lyrics)
```

```python
def get_chief_keef_verses(cleaned_lyrics):
    # Pattern to identify any sections denoted by brackets containing "Chief Keef"
    pattern = re.compile(r"\[.*?chief keef.*?\](.*?)(?=\[|$)", re.DOTALL)
    
    # Find all matches and store them
    chief_keef_verses = pattern.findall(cleaned_lyrics)
    
    # Clean up the verses by stripping unnecessary whitespace
    chief_keef_verses = [verse.strip() for verse in chief_keef_verses]
    
    return chief_keef_verses
```

```python
# Testing verse section
chief_keef_verses = get_chief_keef_verses(cleaned_lyrics)

# Printing
for verse in chief_keef_verses:
    print(verse)
    print("------")
```

```python
# Splitting each verse

def split_verse(verse, split_percent):
    # Tokenize the verse into words
    words = verse.split()
    
    # Calculate the split index
    split_index = round(len(words) * (split_percent / 100))
    
    # Generate the sequence_in and sequence_out
    sequence_in = " ".join(words[:split_index])
    sequence_out = " ".join(words[split_index:])
    
    return sequence_in, sequence_out

# Example usage
verse = "young chop on the beat i can't spell sober i'm a smoker, fredo a drinker, tadoe off molly water sober, my bitches stay sober sober damn, i hate bein' sober i hate bein' sober"
split_percent = 50  # This means 50% of the words will be in sequence_in

sequence_in, sequence_out = split_verse(verse, split_percent)

print("Sequence In (50%):", sequence_in)
print("Sequence Out (50%):", sequence_out)
```

```python
# List to hold all verses by Chief Keef
all_lyrics_by_verse = []

for i, song in enumerate(data['songs']):
    lyrics = song.get('lyrics', None)
    if not lyrics:
        print(f"No lyrics found for song index {i}")
        continue
    cleaned_lyrics = clean_lyrics(lyrics)
    if not cleaned_lyrics:
        print(f"Lyrics are empty after cleaning for song index {i}")
    verses = get_chief_keef_verses(cleaned_lyrics)
    if not verses:
        print(f"No verses extracted for song index {i}")
    all_lyrics_by_verse.extend(verses)

# # Process each song
# for song in data['songs']:
#     lyrics = song['lyrics']  # Assuming each song is a dictionary with a 'lyrics' key
#     cleaned_lyrics = clean_lyrics(lyrics)
#     verses = get_chief_keef_verses(cleaned_lyrics)
#     all_lyrics_by_verse.extend(verses)  # Append each verse found to the list

# # Now `all_lyrics_by_verse` contains all the verses processed
# print("Total verses extracted:", len(all_lyrics_by_verse))
```
