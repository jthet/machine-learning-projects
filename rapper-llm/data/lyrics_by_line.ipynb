```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
file_path = 'chiefKeef_lyrics.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(data.head())
```

```python
# Assuming 'store_lines' is already defined as follows:
def store_lines(text):
    lines = text.split('\n')
    filtered_lines = [line.strip() for line in lines if line.strip()]
    return filtered_lines
```

```python
# Initialize an empty list to store all lines from all songs
all_lines = []

# Iterate over each item in the 'lyrics' column of the DataFrame
for lyrics in data['lyrics']:
    # Get the list of lines for the current song
    song_lines = store_lines(lyrics)
    # Extend the master list with the lines from this song
    all_lines.extend(song_lines)

# Now 'all_lines' contains all the lines from all songs in the DataFrame
print(all_lines)  # You can remove or comment out this print statement in production
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Assume `lyrics` is a list of lyric strings
inputs = tokenizer(all_lines, padding=True, truncation=True, return_tensors="pt")
```
